```python
#importing libraries\
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
#importing dataset
dataset=pd.read_csv('insurance_data.csv')
```

```python
dataset.head(5)
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
#train aur test karenge
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#model banayenge
from sklearn.linear_model import LogisticRegression
```

```python
regressor=LogisticRegression()
regressor.fit(x_train,y_train)
```

```python
y_prediction=regressor.predict(x_test)
```

```python
print(regressor.predict([[25]]))
```

```python
#We will check the how many are correctly predicted through confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_prediction)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_prediction))
```

```python
grouped = dataset.groupby('bought_insurance')['age'].mean()
grouped.plot(kind='bar')
plt.xlabel('bought_insurance')
plt.ylabel('age')
plt.title('age vs bought_insurance ')
plt.show()
```

```python
regressor.score(x_test,y_test)
```
